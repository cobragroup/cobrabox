# SeizureExplorer — interactive demo

This notebook shows how to load a cobrabox `Data` object and launch the
**SeizureExplorer** interactive app inline inside a Jupyter notebook.

The explorer provides three synchronised views:
- **Top** — full-length offset channel traces
- **Middle** — windowed 2-D heatmap
- **Bottom** — mean signal navigator with optional event markers

Use the sidebar controls to navigate position, change the window size, pick
dimensions, adjust the colormap, and jump between events.

```python
import numpy as np
import panel as pn
pn.config.throttled = True
```

## Use a built-in dummy dataset

The `dummy_noise` dataset ships with cobrabox.

```python
from cobrabox.dataset_loader import load_noise_dummy

dataset = load_noise_dummy()
print(f"Dataset has {len(dataset)} recordings")

data = dataset[4]
print(data)
```

## Launch the explorer

`SeizureExplorer(data)` renders inline.  Optional keyword arguments:

| Argument | Default | Description |
|---|---|---|
| `position` | `0` | Initial index of the position marker |
| `window_size` | `200` | Number of samples shown in the heatmap window |
| `point_positions` | `None` | 1-D array of event indices to mark in the navigator |

```python
# Fake seizure-onset events at roughly 1/4, 1/2 and 3/4 of the signal.
n_samples = data.data.sizes[data.data.dims[-1]]
events = np.array([n_samples // 4, n_samples // 2, 3 * n_samples // 4])
```

```python
from cobrabox.visualization import SeizureExplorer

explorer = SeizureExplorer(data, position=events[0], window_size=300, point_positions=events)
explorer
```
